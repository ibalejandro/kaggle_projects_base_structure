# Training plots

```python
from six.moves import cPickle as pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
%matplotlib inline
plt.style.use(["ggplot", "seaborn-ticks"])
```

```python
EXPERIMENT_NAME = "experiment_name"
```

```python
train_history_path = "../../models/{}/history.pickle".format(EXPERIMENT_NAME)
history_plots_path = "./{}/".format(EXPERIMENT_NAME)
os.mkdir(history_plots_path) if not os.path.exists(history_plots_path) else None
with open(train_history_path, "rb") as history_file:
    train_history = pickle.load(history_file)
    print("Keys: {}".format(train_history.keys()))
    epochs = len(train_history["loss"])
    print("Epochs: {}".format(epochs))
```

```python
xticks = list(range(epochs))
xticklabels = [xtick + 1 for xtick in xticks]
print("xticks:")
print(xticks)
print("xticklabels:")
print(xticklabels)
```

```python
# Summarize history for loss.
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(train_history["loss"], marker='^')
ax.plot(train_history["val_loss"], marker='s')
plt.title("Model loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)
plt.legend(["train", "validation"], loc="upper right")
plt.savefig("{}loss.png".format(history_plots_path))
plt.show()
```

```python
# Summarize history for accuracy.
fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(train_history["acc"], marker='^')
ax.plot(train_history["val_acc"], marker='s')
plt.title("Model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)
plt.legend(["train", "validation"], loc="upper left")
plt.savefig("{}accuracy.png".format(history_plots_path))
plt.show()
```
